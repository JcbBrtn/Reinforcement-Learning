```python
from Linear import Linear
import gym
import numpy as np
env = gym.make("LunarLander-v2")
env.action_space.seed(42)
obs, info = env.reset(seed=42, return_info=True)
```

```python
obs
```

```python
#Create everything we need for training
agent = Linear([9])
done = False
round = 0
total_rounds=1
X = np.array([])
Y = []
```

```python
X
```

```python
def get_action(obs, agent):
    expected_rewards = []
    for a in range(4):
        x = np.append(obs, a)
        expected_rewards.append(agent.predict(np.array([x])))
    
    max_r = -1000
    max_i = 0
    for i, r in enumerate(expected_rewards):
        if r[0] > max_r:
            max_r = r
            max_i = i
    
    return max_i
```

```python
X = np.array([])
o = np.array([1,2,3,4])
np.append([X], [o], axis=0)
```

```python
next_act = 0

while round < total_rounds:

    act = get_action(obs, agent)

    new_obs, reward, done, info = env.step(act)
    print(f' Reward: {reward} \n Done: {done} info {info}')
    print(f'Action: {act}')
    env.render()

    #Update X and Y with [obs,action] and reward
    obs = np.append(obs, np.array(act))
    X = np.append(X, obs, axis=0)

    #Update all previous rewards with the current reward
    for i in range(len(Y)):
        Y[i] += reward

    Y = np.append(Y, reward)

    obs = new_obs

    if done:
        obs, info = env.reset(return_info=True)
        round += 1
        #End of the round, do some learning!
        agent.fit(X, Y, 20)

        X = np.array([obs])
        Y = []

env.close()
```
